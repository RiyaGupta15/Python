```python
people = {
    'first' : ['Riya', 'Corey', 'Jane'],
    'last' : ['Gupta', 'Doe', 'Goel'],
    'email' : ['riya.gupta@gmail.com', 'corey.doe@gmail.com', 'jane.goel@gmail.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['email']
```

```python
df.set_index('email')
```

```python
df #Dataframe is still the original, didnt change
```

```python
##To change the original dataframe, add inplace
df.set_index('email', inplace=True)
```

```python
df
```

```python
df.index
```

```python
#Displays info of this particular email coz email is now the index
df.loc['corey.doe@gmail.com'] 
```

```python
df.loc['corey.doe@gmail.com', 'last'] #displays only last name of that email
```

```python
df.iloc[0]  #Gives error with loc coz 0 is not index anymore
```

```python
##Getting back the index values
df.reset_index(inplace=True)
```

```python
df
```
