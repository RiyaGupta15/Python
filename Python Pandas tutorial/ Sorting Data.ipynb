```python
people = {
    'first' : ['Riya', 'Corey', 'Jane', 'Adam'],
    'last' : ['Gupta', 'Doe', 'Doe', 'Doe'],
    'email' : ['riya.gupta@gmail.com', 'corey.doe@gmail.com', 'jane.doe@gmail.com', 'adam.doe@gmail.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.sort_values(by='last')
```

```python
df.sort_values(by='last', ascending=False)
```

```python
#If last names are same, it will sort according to first name
df.sort_values(by=['last', 'first'], ascending=False)
```

```python
#Last names in descending order but first names in ascending order
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
df
```

```python
#Sorting back according to index
df.sort_index()
```

```python
df['last'].sort_values
```
